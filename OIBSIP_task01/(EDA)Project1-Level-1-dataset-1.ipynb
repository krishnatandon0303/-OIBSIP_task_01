LOAD THE RETAIL SALES DATA SET

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#download the dataset from the provided link
#load the date set into jupiter notebook using pandas

'''read csv file and change the date DATATYPE  
from str to timestamp and set date as an index'''

df = pd.read_csv(r"C:\Users\ktand\Downloads\40811eyack.com.MAIL_xsbsxxypt8dh6!App\archive\retail_sales_dataset.csv",parse_dates = ["Date"],index_col= "Date")

df
```

```python
# Assuming your DataFrame is named 'df'
newColumnNames = {
    'Transaction ID': 'TransactionID',
    'Customer ID': 'CustomerID',
    'Gender': 'Gender',
    'Age': 'Age',
    'Product Category': 'ProductCategory',
    'Quantity': 'Quantity',
    'Price per Unit': 'PricePerUnit',
    'Total Amount': 'TotalAmount',
    'Date': 'Date'
}

df.rename(columns = newColumnNames, inplace=True)
df
```

# Calculating basis statistics

Mean,Median,mode and descriptive stats


MEAN

```python
df2= df['Quantity'].mean()
print("The mean of Quantity is :",df2)
print()
df2= df['PricePerUnit'].mean()
print("The mean of PricePerUnit is :",df2)
print()
df2= df['TotalAmount'].mean()
print("The mean of TotalAmount is :",df2) 
print()
```

MEDIAN

```python
median  = df['Quantity'].median()
print("The median of Quantity is :",median)
print()
median  = df['PricePerUnit'].median()
print("The median of PricePerUnit is :",median)
print()
median  = df['TotalAmount'].median()
print("The median of TotalAmount is :",median)
print()
```

MODE

```python
modeValues= df['Quantity'].mode().iloc[0]
print("The mode of Quantity is :",modeValues)
print()
modeValues = df['PricePerUnit'].mode().iloc[0]
print("The mode of PricePerUnit is :",modeValues)
print()
modeValues= df['TotalAmount'].mode().iloc[0]
print("The mode of TotalAmount is :",modeValues)
print()
```

```python
stand_deviation= df['Quantity'].std()
print("The standard deviation of Quantity is :",stand_deviation)
print()
stand_deviation= df['PricePerUnit'].std()
print("The standard deviation of Price_per_Unit is :",stand_deviation)
print()
stand_deviation= df['TotalAmount'].std()
print("The standard deviation of TotalAmount is :",stand_deviation)
print()
```

# Time Analysis Series

Resample the data to monthly frequency and sum the TotalAmount

MONTHLY SALES TREND (MEAN)

```python
monthlySales = df.resample('ME')['TotalAmount'].sum() #total sum of the monthly sales
plt.figure(figsize=(10, 6))
plt.plot(monthlySales, marker='o')
plt.title('MONTHLY SALES TREND')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
```

# CUSTOMER AND PRODUCT ANALYSIS

ANALYSING GENDER DISTRIBUTION

```python
genderCount = df['Gender'].value_counts()

#plotting graph for GENDER DISTRIBUTION
plt.figure(figsize=(8, 5))

plt.bar(genderCount.index, genderCount.values, color='purple',edgecolor='red')
plt.title('GENDER DISTRIBUTION OF CUSTOMERS')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

ANALYSING AGE DISTRIBUTION

```python
ageCount = df['Gender'].value_counts()

#plotting graph for GENDER DISTRIBUTION
plt.figure(figsize=(10, 6))

plt.hist(df['Age'], bins=20, color='blue', edgecolor='red')
plt.title('AGE DISTRIBUTION OF CUSTOMERS')
plt.xlabel('AGE')
plt.ylabel('Count')
plt.show()
```

# CUSTOMER AND PRODUCT ANALYSIS

Sales of products by categorie

```python
categoryTime = df.pivot_table(index=df.index.month, columns='ProductCategory', values='TotalAmount', aggfunc='sum')


plt.figure(figsize=(15, 10))
plt.imshow(categoryTime, cmap='YlGnBu', aspect='auto', interpolation='nearest')
plt.colorbar(label='Total Sales')
plt.title('MONTHLY PRODUCT CATEGORY SALES BY Heatmap')
plt.xlabel('Product Category')
plt.ylabel('Month')

#seetting of ticks in the graph
plt.xticks(range(len(categoryTime.columns)), categoryTime.columns, rotation=45) 
plt.yticks(range(1, 13), range(1, 13))

plt.show()
```
