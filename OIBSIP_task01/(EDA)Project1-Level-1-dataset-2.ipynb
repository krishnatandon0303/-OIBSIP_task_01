# MENU DATASET

LOAD THE FIRST DATASET

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#load the file path using read csv 

df = pd.read_csv(r"C:\Users\ktand\Downloads\menu.csv")
df
```

# RENAMING THE DATASET COLUMNS

```python

new_column_names = {
    'Category': 'Category',
    'Item': 'Item',
    'Serving Size': 'ServingSize',
    'Calories': 'Calories',
    'Calories from Fat': 'CaloriesFromFat',
    'Total Fat': 'TotalFat',
    'Total Fat (% Daily Value)': 'TotalFatPercentDV',
    'Saturated Fat': 'SaturatedFat',
    'Saturated Fat (% Daily Value)': 'SaturatedFatPercentDV',
    'Trans Fat': 'TransFat',
    # ... (add other column mappings here)
    'Carbohydrates': 'Carbohydrates',
    'Carbohydrates (% Daily Value)': 'CarbohydratesPercentDV',
    'Dietary Fiber': 'DietaryFiber',
    'Dietary Fiber (% Daily Value)': 'DietaryFiberPercentDV',
    'Sugars': 'Sugars',
    'Protein': 'Protein',
    'Vitamin A (% Daily Value)': 'VitaminAPercentDV',
    'Vitamin C (% Daily Value)': 'VitaminCPercentDV',
    'Calcium (% Daily Value)': 'CalciumPercentDV',
    'Iron (% Daily Value)': 'IronPercentDV'
}

df.rename(columns=new_column_names, inplace=True)
df
```

CREATING DUMMY DATES FOR DATASET

```python
num_rows = len(df)
start_date = pd.to_datetime('2023-01-01')
date_range = pd.date_range(start_date, periods=num_rows, freq='D')
df['Date'] = date_range

# Convert 'Date' to datetime format (if needed)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Set 'Date' as the index for time series analysis
df.set_index('Date', inplace=True)
df
```

# CLEANING THE DATASET

```python
# find the duplicates values in the dataset
df2 = df.duplicated()
df2
#deleting the duplicates
df3= df.drop_duplicates()
df3
```

# TIME SERIES ANALYSIS 

RESAMAPLING


```python
# Resample data by month and calculate total sales (using 'Calories' as an example metric)
monthlySales = df['Calories'].resample('ME').sum()

# Plot time series

plt.figure(figsize=(10, 6))
plt.plot(monthlySales)
plt.title('Monthly Calories Over Time')
plt.xlabel('Date')
plt.ylabel('Total Calories')
plt.show()
```

BAR CHARTS OF CALORIES BY CATEGORY

```python

df_grouped = df.groupby('Category').sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(df_grouped['Category'], df_grouped['Calories'],color = 'green',edgecolor ='black')
plt.title('Total Calories by Category')
plt.xlabel('Category')
plt.ylabel('Total Calories')
plt.xticks(rotation=90)
plt.show()
```

Select only numeric columns for correlation

Heatmap of nutritional information correlations


```python
# Select only numeric columns for correlation
numeric_cols = df.select_dtypes(include=[np.number])

# Heatmap of nutritional information correlations
plt.figure(figsize=(12, 9))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
